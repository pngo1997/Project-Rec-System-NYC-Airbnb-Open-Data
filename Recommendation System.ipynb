**Calculate compound rating score based on Ratings Negative-Neutral-Postive. First option: Scale the original compound scores to a normalized range of 1-5.**

```python
scoreWeights = {'Rating Negative': 1, 'Rating Neutral': 3, 'Rating Positive': 5}
compoundScore = ratings[['Rating Negative', 'Rating Neutral', 'Rating Positive']].mul(pd.Series(scoreWeights)).sum(axis=1)
#Multiply each ratings with corresponding weights and sum up.

#Scale the compound scores to a normalized range of 1-5.
rating1 = compoundScore.apply(lambda x: round((x - compoundScore.min()) / (compoundScore.max() - compoundScore.min()) * 4 + 1))

#Calculate the frequency of each ranking value.
ratingFrequency1 = rating1.value_counts().sort_index()
print(ratingFrequency1)
```

**Calculate compound ratig score based on Ratings Negative-Neutral-Postive. Second option: Rounding the original compound score to the nearest integer value.**

```python
scoreWeights = {'Rating Negative': 1, 'Rating Neutral': 3, 'Rating Positive': 5}

compoundScore = ratings[['Rating Negative', 'Rating Neutral', 'Rating Positive']].mul(pd.Series(scoreWeights)).sum(axis=1)
#Multiply each ratings with corresponding weights and sum up.

#Rounding the compound score to the nearest integer value. 
rating2 = compoundScore.apply(lambda x: round(x))

#Calculate the frequency of each ranking value.
ratingFrequency2 = rating2.value_counts().sort_index()
print(ratingFrequency2)
```

## **class for Recommendation system.**

```python
class RecommendationSystem:
    def __init__(self, data, vectorizer):
        self.data = data
        self.vectorizer = vectorizer

    def preprocessText(self, text):
        '''Return text in lowercase.'''
        preprocessedText = text.lower()
        return preprocessedText

    def recommendListings(self, user_price, user_neighborhood, user_rating, topRecs=10):
        recommendData = self.data.copy()

        if user_price and user_price != "":
            user_price = float(user_price)
            recommendData = recommendData[recommendData['price'] <= user_price]

        if user_neighborhood and user_neighborhood != "":
            recommendData = recommendData[recommendData['neighbourhood_group'].apply(self.preprocessText) == self.preprocessText(user_neighborhood)]

        if user_rating and user_rating != "":
            user_rating = float(user_rating)
            recommendData = recommendData[recommendData['Rating'] >= user_rating]

        #Create a user input vector.
        user_input = [user_price, user_neighborhood, user_rating]
        user_input_vector = self.vectorizer.transform([self.preprocessText(" ".join(str(x) for x in user_input))])

        #Calculate the similarity between the user input and listings features.
        listing_features = recommendData[['price', 'neighbourhood_group', 'Rating']]
        listing_feature_vectors = self.vectorizer.transform(listing_features['neighbourhood_group'].apply(self.preprocessText))
        similarities = cosine_similarity(user_input_vector, listing_feature_vectors).flatten()

        #Sort the listings based on similarity in descending order.
        indices = similarities.argsort()[::-1]
        recommended_listings = recommendData.iloc[indices].head(topRecs)

        #Display the top recommended listings.
        print(f"Top {topRecs} recommended listings:")
        for _, listing in recommended_listings.iterrows():
            listingID = listing['id']
            stationDistance = listing['Distance_from_station']
            stationName = listing['Station_name']
            Airbnb_type = listing['Airbnb_type']
            print(f"- Listing ID: {listingID}, Distance to Nearest Subway Station: {stationDistance:.3f} miles, Nearest Station: {stationName}, Airbnb Type : {Airbnb_type}")
```

```python
recSys = RecommendationSystem(data, vectorizer)

userPrice = input("Enter price: ")
userNeighborhood = input("Enter neighborhood group: ")
userRating = input("Enter rating: ")

#Recommend listings based on user inputs
recSys.recommendListings(userPrice, userNeighborhood, userRating)
```
